### **Yellowstone Park**


Yellowstone abriga a maior concentração de mamíferos nos 48 estados contíguos dos EUA. O parque é notável por seu complexo predador-presa envolvendo grandes mamíferos. O objetivo do Serviço Nacional de Parques é manter os processos ecológicos que sustentam esses mamíferos e seus habitats, enquanto monitora as mudanças em suas populações.


**Lobo**


Os lobos são animais sociais que vivem em matilhas, com tamanho médio de 11,8 indivíduos em Yellowstone. As matilhas são famílias complexas com membros que possuem papéis e personalidades distintas. Eles defendem seu território com uivos e marcas de cheiro. A vida social dos lobos oferece vantagens como caça cooperativa, cuidado dos filhotes e defesa do território.Os lobos de Yellowstone caçam uma variedade de presas: alces, castores, coelhos etc.

**Castor**

Os castores (Castor canadensis) são uma espécie fundamental em Yellowstone, moldando o ambiente através da construção de represas e da derrubada de árvores. Eles preferem habitats com salgueiros, álamos e choupos, concentrando-se em áreas com esses recursos. A população de castores no parque variou ao longo dos anos, com 25 colônias registradas em 1921 e um aumento para 108 colônias em 2021, graças à regeneração do salgueiro e reintroduções de castores nas florestas nacionais ao norte. A população atual reflete tanto flutuações naturais quanto melhorias na visibilidade das colônias durante pesquisas aéreas.

**Coelho**

O coelho, tem uma pelagem cinza-marrom que se torna quase branca no inverno para camuflagem na neve. Em Yellowstone, a maioria das 501 observações feitas em 2008 ocorreu em habitats de arbustos de sálvia a altitudes abaixo de 6.500 pés. Porém com a forte predação por parte de animais como lobos, e competições diretas entre espécies, fez com que houvesse uma queda na população de coelhos em Yellowstone.

### **Modelo**

O modelo Lotka-Volterra tradicionalmente descreve a dinâmica entre duas espécies interagindo em um ecossistema, como uma espécie predadora e uma espécie presa. No entanto, em ecossistemas mais complexos, é comum encontrar interações envolvendo mais de duas espécies. Esse é o caso do modelo trifásico, onde três espécies interagem de maneira direta e indireta. O estudo dessas interações é uma extensão natural do trabalho de Alfred J. Lotka e Vito Volterra, que estabeleceram as bases para a modelagem matemática de ecossistemas.

A aplicação do modelo Lotka-Volterra em um sistema trifásico busca capturar as complexas relações entre predadores, presas e outras espécies que podem atuar como competidores, mutualistas ou ter algum outro tipo de interação. No caso do ecossistema de Lotka-Volterra, estamos interessados nas interações entre lobos (predadores), coelhos (presas) e castores (competidores).

### **Visualização**

Cada espécie, (note que é a população de cada espécie separadamente, porém ainda inserida no contexto da competição e predação trifásico).

```python
import numpy as np
from matplotlib.pylab import *

# Parâmetros do modelo, Observados de acordo com Yellowstone Park
A, B, C, M, N, V, D, K, G, T = [1.75, 0.0015, 0.0008, 1.1, 0.0015, 0.0001, 0.0015, 0.0015, 0.8, 1.2]
#A = a taxa de crescimento dos Coelhos na ausência de predadores;
#B = a taxa de predação do Coelho pelo Lobo;
#C = a taxa de competição por recursos do Coelho com o Castor;
#M = a taxa de crescimento dos Castores na ausência de predadores;
#N = a taxa de predação do Castor pelo Lobo;
#V = a taxa de competição por recursos entre castores e coelhos;
#D = a eficiência de conversão dos Coelhos em novos Lobos;
#K = a eficiência de conversão dos Castores em novos lobos;
#G = a taxa de mortalidade do Lobo na ausência de Coelhos;
#T = a taxa de mortalidade do Lobo na ausência de Castores;

def f1(L, Clh, Cst):  # Equação para Coelhos
    return (A * Clh) - (B * Clh * L) - (C * Clh * Cst)

def f2(L, Clh, Cst):  # Equação para Castores
    return (M * Cst) - (N * Cst * L) - (V * Clh * Cst)

def f3(L, Clh, Cst):  # Equação para Lobos
    return (D * L * Clh + K * L * Cst) - (G * L) - (T * L)

# Condições iniciais de acordo com Yellowstone Park
L0 = 110  # Lobos
Clh0 = 501  # Coelhos
Cst0 = 708  # Castores
T_total = 20  # Tempo total de simulação
n_steps = 1000  # Número de passos

# Inicialização dos arrays para armazenar os resultados
L = np.zeros(n_steps + 1)
Clh = np.zeros(n_steps + 1)
Cst = np.zeros(n_steps + 1)
L[0], Clh[0], Cst[0] = L0, Clh0, Cst0

# Simulação usando o método de Euler
dt = T_total / n_steps
for k in range(n_steps):
    L[k+1] = max(0, L[k] + dt * f3(L[k], Clh[k], Cst[k]))
    Clh[k+1] = max(0, Clh[k] + dt * f1(L[k], Clh[k], Cst[k]))
    Cst[k+1] = max(0, Cst[k] + dt * f2(L[k], Clh[k], Cst[k]))
```

```python
plt.plot(np.linspace(0, T_total, n_steps + 1), Clh, label='Coelhos')
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Modelo de Presa-Predador com Competição')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(np.linspace(0, T_total, n_steps + 1), Cst, label='Castores')
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Modelo de Presa-Predador com Competição')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(np.linspace(0, T_total, n_steps + 1), L, label='Lobos')
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Modelo de Presa-Predador com Competição')
plt.legend()
plt.grid(True)
plt.show()
```

Com 2 espécies, (note que é a população de cada espécie em duplas, porém ainda inserida no contexto da competição e predação trifásico).

Encontrando o valor mínimo de oscilação populacional de cada espécie

```python
min_lobos = np.min(L)
min_castores = np.min(Cst)
min_coelhos = np.min(Clh)

print("{:.0f}".format(min_lobos))
print("{:.0f}".format(min_castores))
print("{:.0f}".format(min_coelhos))
```
